```python
# 1. Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Create a NumPy array and perform basic operations
arr = np.array([1, 2, 3, 4, 5])
print('Array:', arr)
print('Mean:', np.mean(arr))
print('Sum:', np.sum(arr))
```

```python
# 3. Generate a random NumPy array and compute statistics
random_arr = np.random.rand(5, 5)
print('Random Array:\n', random_arr)
print('Standard Deviation:', np.std(random_arr))
```

```python
# 4. Create a Pandas DataFrame and display basic information
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})
print(df.info())
print(df.describe())
```

```python
# 6. Plot a simple line graph using Matplotlib
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# 7. Create a Seaborn histogram
sns.histplot(data=np.random.randn(1000), bins=30, kde=True)
plt.title('Histogram with KDE')
plt.show()
```

```python
# 8. Perform a groupby operation in Pandas
df = pd.DataFrame({'Category': ['A', 'B', 'A', 'B'], 'Values': [10, 20, 30, 40]})
grouped = df.groupby('Category').sum()
print(grouped)
```

```python
# 9. Create a scatter plot using Seaborn
df = pd.DataFrame({'x': np.random.rand(50), 'y': np.random.rand(50)})
sns.scatterplot(x='x', y='y', data=df)
plt.title('Random Scatter Plot')
plt.show()
```

```python
# 10. Generate a heatmap using Seaborn
data = np.random.rand(10, 10)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
